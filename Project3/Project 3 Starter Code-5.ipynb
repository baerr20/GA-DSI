```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
df=pd.read_csv('/Users/Marina/Documents/GA Data Science/projects/projects-weekly/project-03/model_frame.csv')
```

### Now we will be building or model by adding our predictors step-wise. If our MSE is decreasing then the predictor is improving out model's fit. It is also important whether our predictors are statistically siginificant. 

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error



model0 = smf.ols(formula = 'Sale ~ Retail', data = df).fit()
predictions0 = model0.predict()


plt.scatter(predictions0, df["Sale"])
plt.xlabel("Predicted Values")
plt.ylabel("Sale")
plt.show()
print "MSE:", mean_squared_error(df["Sale"], predictions0)
print(model0.summary())


# adding Sold decreased our MSE so we can leave this predictor in our model 
model1 = smf.ols(formula = 'Sale ~ Retail + Sold', data = df).fit()
predictions1 = model1.predict()

plt.scatter(predictions1, df["Sale"])
plt.xlabel("Predicted Values")
plt.ylabel("Sale")
plt.show()
print "MSE:", mean_squared_error(df["Sale"], predictions1)
print(model1.summary())

# adding Store (number of stores per zip code) decreased our MSE so we can leave this predictor in our model 
model2 = smf.ols(formula = 'Sale ~ Retail + Sold + Store', data = df).fit()
predictions2 = model2.predict()

plt.scatter(predictions2, df["Sale"])
plt.xlabel("Predicted Values")
plt.ylabel("Sale")
plt.show()
print "MSE:", mean_squared_error(df["Sale"], predictions2)
print(model2.summary())

# adding Vendor (number of vendors per zip code) did not change our MSE (although, it was significant 
#predictor of sale) so we decided not to include it in our model since there was no change in the validation score.
model3 = smf.ols(formula = 'Sale ~ Retail + Sold + Store + Vendor', data = df).fit()
predictions3 = model3.predict()

plt.scatter(predictions3, df["Sale"])
plt.xlabel("Predicted Values")
plt.ylabel("Sale")
plt.show()
print "MSE:", mean_squared_error(df["Sale"], predictions3)
print(model3.summary())


# adding Month (the month when the sales transaction took place) decreased our MSE so we can leave this predictor in our model 

model4 = smf.ols(formula = 'Sale ~ Retail + Sold + Store + C(Month)', data = df).fit()
predictions4 = model4.predict()

plt.scatter(predictions4, df["Sale"])
plt.xlabel("Predicted Values")
plt.ylabel("Sale")
plt.show()
print "MSE:", mean_squared_error(df["Sale"], predictions4)
print(model4.summary())


# adding Volume (the average volume per bottle) decreased our MSE so we can leave this predictor in our model 

model5 = smf.ols(formula = 'Sale ~ Retail + Sold + Store + C(Month) + Volume', data = df).fit()
predictions5 = model5.predict()

plt.scatter(predictions5, df["Sale"])
plt.xlabel("Predicted Values")
plt.ylabel("Sale")
plt.show()
print "MSE:", mean_squared_error(df["Sale"], predictions5)
print(model5.summary())
```

### Our final model includes Month, Retail, Sold, Store, and Volume for the predictors and the average sale per transaction for the dependent variable. Now, we are going to cross-validate our model. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
df=pd.read_csv('/Users/Marina/Documents/GA Data Science/projects/projects-weekly/project-03/model_frame.csv')
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error



#fitting model to a small random subset

subset = df.sample(frac=0.1, replace=False)

model0 = smf.ols(formula = 'Sale ~ Retail + Sold + Store + C(Month) + Volume', data = subset).fit()
predictions0 = model0.predict()

plt.scatter(predictions0, subset["Sale"])
plt.xlabel("Predicted Values")
plt.ylabel("Sale")
plt.show()
print "Sample1:", mean_squared_error(subset["Sale"], predictions0)
print model0.summary()

#fitting model to a bit larger random subset

subset1 = df.sample(frac=0.4, replace=False)

model1 = smf.ols(formula = 'Sale ~ Retail + Sold + Store + C(Month) + Volume', data = subset1).fit()
predictions1 = model1.predict()

plt.scatter(predictions1, subset1["Sale"])
plt.xlabel("Predicted Values")
plt.ylabel("Sale")
plt.show()
print "Sample2:", mean_squared_error(subset1["Sale"], predictions1)
print model1.summary()

#fitting model to the whole data set

model2 = smf.ols(formula = 'Sale ~ Retail + Sold + Store + C(Month) + Volume', data = df).fit()
predictions2 = model2.predict()

plt.scatter(predictions2, df["Sale"]) 
plt.xlabel("Predicted Values")
plt.ylabel("Sale")
plt.show()
print "All data with sample model:", mean_squared_error(df["Sale"], predictions2)
print model2.summary()


#fitting model to a small random subset with regulization

subset = df.sample(frac=0.1, replace=False)

model3 = smf.ols(formula = 'Sale ~ Retail + Sold + Store + C(Month) + Volume', data = subset).fit_regularized(L1_wt=0.0)
predictions3 = model3.predict()

plt.scatter(predictions3, subset["Sale"])
plt.xlabel("Predicted Values")
plt.ylabel("Sale")
plt.show()
print "Sample1 with regulization:", mean_squared_error(subset["Sale"], predictions3)
print model3.summary()

#fitting model to a bit larger random subset with regulization

subset1 = df.sample(frac=0.5, replace=False)

model4 = smf.ols(formula = 'Sale ~ Retail + Sold + Store + C(Month) + Volume', data = subset1).fit_regularized(L1_wt=0.0)
predictions4 = model4.predict()

plt.scatter(predictions4, subset1["Sale"])
plt.xlabel("Predicted Values")
plt.ylabel("Sale")
plt.show()
print "Sample2 with regulization:", mean_squared_error(subset1["Sale"], predictions4)
print model4.summary()

#fitting model to the whole data set with regulization

model5 = smf.ols(formula = 'Sale ~ Retail + Sold + Store + C(Month) + Volume', data = df).fit_regularized(L1_wt=0.0)
predictions5 = model5.predict()

plt.scatter(predictions5, df["Sale"]) 
plt.xlabel("Predicted Values")
plt.ylabel("Sale")
plt.show()
print "All data with sample model with regulization:", mean_squared_error(df["Sale"], predictions5)
print model5.summary()


```
