```python
# function for getting mountain name from snow report

def extract_mountain(loc):
    try:
        for n in loc.findAll(class_='name'):
            return n.text.strip()
    except:
        return 'None'
```

```python
# function for getting day snow forecast

def extract_day_snow_forecast(snow):
    try:
        for n in snow.findAll(class_='fblock day nsnow'):
            return n.text.strip()
    except:
        return 'None' 
```
